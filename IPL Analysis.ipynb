```python
#Match Details
query="""
SELECT Match_Id,Season_Id, Home_Team.Team_Name AS Home_Team_Name, Away_Team.Team_Name AS Away_Team_Name, Win_Team.Team_Name AS Winner, Win_Margin, Win_By.Win_Type as Win_Type, Player_Name AS Man_Of_The_Match
FROM Match
INNER JOIN Team as Home_Team ON Match.Team_1=Home_Team.Team_Id
INNER JOIN Team as Away_Team ON Match.Team_2=Away_Team.Team_Id
INNER JOIN Team as Win_Team ON Match.Match_Winner=Win_Team.Team_Id
INNER JOIN Win_By ON Win_By.Win_Id=Match.Win_Type
INNER JOIN Player ON Player.Player_Id= Match.Man_of_the_Match
"""

match=pd.read_sql(query, conn)
match
```

```python
#Maximum Man of the Match Award

query="""
WITH matches AS (
SELECT Match_Id,Season_Id, Home_Team.Team_Name AS Home_Team_Name, Away_Team.Team_Name AS Away_Team_Name, Win_Team.Team_Name AS Winner, Win_Margin, Win_By.Win_Type as Win_Type, Player_Name AS Man_Of_The_Match
FROM Match
INNER JOIN Team as Home_Team ON Match.Team_1=Home_Team.Team_Id
INNER JOIN Team as Away_Team ON Match.Team_2=Away_Team.Team_Id
INNER JOIN Team as Win_Team ON Match.Match_Winner=Win_Team.Team_Id
INNER JOIN Win_By ON Win_By.Win_Id=Match.Win_Type
INNER JOIN Player ON Player.Player_Id= Match.Man_of_the_Match)

SELECT Man_Of_The_Match,COUNT(1) as Number
FROM matches
GROUP BY Man_Of_The_Match
ORDER BY COUNT(1) DESC
"""

mom=pd.read_sql(query, conn)
mom.head()
```
